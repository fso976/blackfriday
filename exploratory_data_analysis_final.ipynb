# Exploratory Data Analysis

```python
# import the libraries
%matplotlib inline

import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

# apply style to all the charts
sns.set_style('whitegrid')

# convert scientific notation to decimals
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Load the clean dataset

```python
# Load the data
df = pd.read_csv('black_friday_processed.csv')
df
```

## Data Perspective

### One Variable (numeric)

```python
# Analize purchase distribuction

# Histogram
print(df['Purchase'].describe().round())
plt.hist(df['Purchase'], bins=100)
plt.show()
# Box plot
plt.boxplot(df['Purchase'])
plt.xticks([1], ['Purchase'], rotation='horizontal')
plt.show()
```

As we can see through the histogram and bosplot charts, the majority of transactions in the dataset results in purchase amounts between 5000 and 15000 dolars, with 75% of the transactions resulting in 12000 purchase amount. The maximum purchase amount is less than 25000 dolars.

## One Variable (categorical)

```python
# Analize gender distribution
print(df['Gender'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchases')

labels = df['Gender'].value_counts().index
values = df['Gender'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

As we can see in this bar chart, more than 75% of the transactions were made by men, while women were responsible for less 25% of them.

```python
# Analize age distribuction
print(df['Age'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by Age')
plt.xlabel('Age Rate')
plt.ylabel('Purchases')

labels = df['Age'].value_counts().index
values = df['Age'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

In this case, the bar chart shows that the costumers between 26 and 35 years old are the main buyers for the retail store (40% of the transactions), while the teenagers until 17 years old are the less expressive ones (3% of the transactions).

```python
# Analize city category distribuction
print(df['City_Category'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by City')
plt.xlabel('City_Category')
plt.ylabel('Purchases')

labels = df['City_Category'].value_counts().index
values = df['City_Category'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

This bar chart shows that the city B concentrates the majority of transactions for the retail store.

## Multiple variables (Categorical x Numeric)

```python
# Visualization : correlation between Product of Category 1 x Purchase
print('Correlation:', df[['Product_Category_1', 'Purchase']].corr())

plt.figure(figsize=(10,10))
plt.xlabel('Product_Category_1')
plt.ylabel('Purchase')

plt.plot(df['Product_Category_1'], df['Purchase'], 'o', alpha=0.3)
plt.show()
```

```python
# Visualization : correlation between Product of Category 2 x Purchase
print('Correlation:', df[['Product_Category_2', 'Purchase']].corr())

plt.figure(figsize=(10,10))
plt.xlabel('Product_Category_2')
plt.ylabel('Purchase')

plt.plot(df['Product_Category_2'], df['Purchase'], 'o', alpha=0.3)
plt.show()
```

```python
# Visualization : correlation between Product of Category 3 x Purchase 
print('Correlation:', df[['Product_Category_3', 'Purchase']].corr())

plt.figure(figsize=(10,10))
plt.xlabel('Product_Category_3')
plt.ylabel('Purchase')

plt.plot(df['Product_Category_3'], df['Purchase'], 'o', alpha=0.3)
plt.show()
```

```python
# Scatterplot Occupation x Purchase
print('Correlation:', df[['Occupation', 'Purchase']].corr())

plt.figure(figsize=(10,10))
plt.xlabel('Occupation')
plt.ylabel('Purchase')

plt.plot(df['Occupation'], df['Purchase'], 'o', alpha=0.3)
plt.show()
```

```python
# Box plot to show the distribution between City_Category x Purchase
print('Correlation:', df[['Marital_Status', 'Purchase']].corr())

values = []
labels = []
for city in sorted(df['Marital_Status'].unique()):
    labels.append(city)
    values.append(list(df[df['Marital_Status']==city]['Purchase'].values))

plt.boxplot(values)
plt.xticks(range(1,len(labels)+1), labels, rotation='horizontal')
plt.show()
```

None of the correlations above are very strong, but we can say that the the most correlated variables are Product_Category_1 and Purchase.

```python
# Cat plot to show the distribution between Occupation x Purchase
sns.catplot(x = "Occupation",
            y = "Purchase",
            hue = "City_Category",
            hue_order = ["A", "B", "C"],
            aspect = 3,
            kind = "bar",
            data = df)
```

This figure accounts for the occupation of our customers across different cities in terms of purchase. Let's first address the spike for Occupation 8 in City A, it showed in the earlier countplot that occupation 8 made up only of a tiny number of purchases, the vertical std bar helps explain this spike probably by a few large transactions. However, we can see that apart from the occupation 8, the graph shows that purchases are highest in City C regardless of occupations. We can therefore make assumptions about the purchasing power from customers in City C.

```python
# Catplot for Age x Purchase considering the Gender and the City Category
sns.catplot(x = 'Age',
           y = 'Purchase',
           hue = 'Gender',
           col = 'City_Category',
           order = ['0-17', '18-25','26-35', '36-45','46-50', '51-55','55+'],
           hue_order = ['M', 'F'],   
           data = df)
```

The above figure shows us the distribution of purchase amounts for respective age group across the cities A, B and C.  Note that in general men in the 26-35 age group are our largest in number and in amount as our highest spending group across all three cities. Women, on the other hand, occurs to peak in number and purchase amount within the 46-50 age group except in City A where the majority of purchase were clearly made by the 0-17 age group.   

Based on the above summary, we can advise the store manager to gear their sales effort on the said gender specific age groups on next black friday. The type of products to be chosen will depend on the store past experiences dealing with the respective age group with products that are known to be popular for specific gender.
