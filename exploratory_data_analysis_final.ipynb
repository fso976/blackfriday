# Exploratory Data Analysis

```python
# import the libraries
%matplotlib inline

import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

# apply style to all the charts
sns.set_style('whitegrid')

# convert scientific notation to decimals
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Load the clean dataset

```python
# Load the data
df = pd.read_csv('black_friday_processed.csv')
df
```

## Data Perspective

### One Variable (numeric)

```python
# Analize purchase distribution

# Histogram
print(df['Purchase'].describe().round())
plt.hist(df['Purchase'], bins=100)
plt.show()
# Box plot
plt.boxplot(df['Purchase'])
plt.xticks([1], ['Purchase'], rotation='horizontal')
plt.show()
```

## One Variable (categorical)

```python
# Analize gender distribution
print(df['Gender'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchases')

labels = df['Gender'].value_counts().index
values = df['Gender'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

```python
# Analize age distribution
print(df['Age'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by Age')
plt.xlabel('Age Rate')
plt.ylabel('Purchases')

labels = df['Age'].value_counts().index
values = df['Age'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

```python
# Analize city category distribution
print(df['City_Category'].value_counts())

# Bar Chart
plt.figure(figsize=(12,6))
plt.title('# Purchases by City')
plt.xlabel('City_Category')
plt.ylabel('Purchases')

labels = df['City_Category'].value_counts().index
values = df['City_Category'].value_counts().values

y_pos = range(len(labels))
plt.bar(y_pos, values, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.show()
```

## Multiple variables (Numeric x Numeric)

```python
# Visualization : correlation between Purchase x Product of Category 1

plt.figure(figsize=(10,10))
plt.xlabel('Purchase')
plt.ylabel('Product_Category_1')

plt.plot(df['Purchase'], df['Product_Category_1'], 'o', alpha=0.3)
plt.show()
```

```python
# Scatterplot Purchase x Occupation
# TODO: What is the correlation between Price x Mean Income

print('Correlation:', df[['Purchase', 'Occupation']].corr())

plt.figure(figsize=(10,10))
plt.xlabel('Purchase')
plt.ylabel('Occupation')

plt.plot(df['Purchase'], df['Occupation'], 'o', alpha=0.3)
plt.show()
```

```python
# Box plot to show the distribution between Purchases x City_Category
values = []
labels = []
for bdr in sorted(df['City_Category'].unique()):
    labels.append(bdr)
    values.append(list(df[df['City_Category']==bdr]['Purchase'].values))

plt.boxplot(values)
plt.xticks(range(1,len(labels)+1), labels, rotation='horizontal')
plt.show()
```
