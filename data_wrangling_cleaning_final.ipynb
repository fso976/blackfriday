### Dealing with outliers

Using Violinplot to find outliers based on City categoryÂ¶

```python
g = sns.catplot(x = "City_Category", 
                y = "Purchase",
                order = ["A", "B", "C"],
                palette = "husl",
                kind = "violin",
                data = df,
                )
```

As we can see in the Violinplot chart, the distribution of purchases between cities is very similar and there is no reason to remove outliers based on that feature.

Using Boxplot to find outliers based on Gender and City Categorie

```python
sns.catplot(x="City_Category",
            y="Purchase",
            hue="Gender",
            hue_order=["M", "F"],
            order=["A", "B", "C"],
            kind="box",
            palette="pastel",
            data=df);
```

```python
# Trying to find outliers in purchases of products of category 1
print(df['Product_Category_1'].value_counts())
print('mean', np.mean(df['Product_Category_1']))
print('std', np.std(df['Product_Category_1']))
plt.hist(df['Product_Category_1'], bins=20)
plt.show()
```

As we can see, there are some products of category one that could be considered outliers, but we don't want to skew the ranking of the product category.

```python
# Calculating the outliers
mean_df = df['Purchase'].mean()
std_df = df['Purchase'].std()

print("Mean Purchase = " + str(mean_df))
print("Standard deviation of purchase = " + str(std_df))
```

To calculate the confidence interval, we used (Mean +/- 3 * std) for upper/lower thresholds:
In our case,
    9333.86 +/- 3(4981) = -5609.14 or 24276.86
    As we can see above, 24276.86 is higher than the highest purchase amount we have in our dataset which is 23961. Therefore, we have decided to keep all the outliers as we prefer to preserve the integrity of our data while all data are well within our acceptable limit. On the other end of the spectrum, we do have any negative amount, thus the lower limit is irrelevant for our analysis.
